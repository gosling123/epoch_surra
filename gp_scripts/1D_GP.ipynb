```python
kernel = 'Exponential'
input_type = 'I'
output_type = 'T_hot'
input_file = '../training_results/train_1D_500.json'
output_file = '../training_results/train_out_1D_500.json'
train_frac = 0.99
restarts = 40
```

```python
epoch_1DGP = GP_1D(kernel, input_type, output_type, input_file, output_file, train_frac, restarts)
```

```python
epoch_1DGP.sacle_data()
```

```python
model = epoch_1DGP.regression()
```

```python
epoch_1DGP.test_train_plot(model)
```

```python
I = np.geomspace(1e14, 1e16, 1000)[:, None]

Y_p, V_p = epoch_1DGP.GP_predict(model, I)

print(np.sqrt(V_p))
```

```python
plt.semilogx(I, Y_p, '-o')
plt.fill_between(I.flatten(), Y_p.flatten(), Y_p.flatten()+2*np.sqrt(V_p.flatten()), color='blue', alpha=0.1, label='Total uncertainty')
plt.fill_between(I.flatten(), Y_p.flatten(), Y_p.flatten()-2*np.sqrt(V_p.flatten()), color='blue', alpha=0.1, label='Total uncertainty')
```
