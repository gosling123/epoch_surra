```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import csv

plt.rcParams["figure.figsize"] = (20,3)
plt.rcParams["figure.figsize"] = [15, 15]
plt.rcParams["figure.autolayout"] = True
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['legend.fontsize'] = 20
```

```python
def average(array, num):
    data = array[:num]
    n,m = data.shape
    av = np.zeros(m)
    for i in range(n):
        av += data[i]
    av /= n
    return av
```

### PPC 100 test (Fig. 1 SJ et al.)

```python
I = np.load('Ensemble_I_test.npy')
I_srs= np.load('Ensemble_SRS_test.npy')
```

```python
I_SJ = np.load('I_L_SJ_data.npy')
I_srs_SJ = np.load('I_SRS_SJ_data.npy')
```

```python
for i in range(10):

    plt.loglog(I[i], I_srs[i], '-o')
    plt.gcf().set_size_inches(10,8)
```

```python
I_srs_av = np.zeros((10,20))
I_av = np.zeros((10,20))

for i in range(len(I_av)):
    I_av[i] = average(I, num=int(i+1))
    I_srs_av[i] = average(I_srs, num=int(i+1))
```

```python
plt.loglog(I_SJ[0], I_srs_SJ[0], '-o', color = 'blue', label = 'SJ et al.')
plt.plot(I_av[-1], I_srs_av[-1], '-o', color = 'black', label = f'{len(I_srs)} simulation average')
plt.xlabel(r'$I_{L} W/cm^2$', fontsize = 16)
plt.ylabel(r'$I_{srs} W/cm^2$', fontsize = 16)
plt.legend(fontsize = 14)
plt.gcf().set_size_inches(10,8)
```

```python
plt.subplot(1,3,1)
for i in range(len(I)):
    plt.loglog(I[i], I_srs[i], alpha = 0.65)
plt.loglog(I_av[-1], I_srs_av[-1], '-o', color = 'black', label = 'Average')
plt.xlabel(r'$I_L \,\, W/cm^2 $')
plt.ylabel(r'$I_srs \,\, W/cm^2 $')
plt.legend()

plt.subplot(1,3,2)
plt.loglog(I_SJ[0], I_srs_SJ[0], '-o', color = 'blue', label = 'SJ et al.')
plt.plot(I_av[-1], I_srs_av[-1], '-o', color = 'black', label = f'Average')
plt.xlabel(r'$I_{L} W/cm^2$')
plt.ylabel(r'$I_{srs} W/cm^2$')
plt.legend()

plt.subplot(1,3,3)
for i in range(len(I)):
    plt.loglog(I_av[i], I_srs_av[i], '-o', label = f'{i+1} Siumlation Average')
plt.xlabel(r'$I_L \,\, W/cm^2 $')
plt.ylabel(r'$I_srs \,\, W/cm^2 $')
plt.legend()

plt.gcf().set_size_inches(30,10)
```
